```python
import math
import copy


def distance(points):
    return math.sqrt((points[0][0] - points[1][0]) *
                     (points[0][0] - points[1][0]) +
                     (points[0][1] - points[1][1]) *
                     (points[0][1] - points[1][1]) +
                     (points[0][2] - points[1][2]) *
                     (points[0][2] - points[1][2]))


def smallest_distance(points, n):
    min_val = float('inf')
    min_points = []
    for i in range(n):
        for j in range(i + 1, n):
            if distance([points[i], points[j]]) < min_val:
                min_val = distance([points[i], points[j]])
                min_points = [points[i], points[j]]

    return min_val, min_points



def strip_closest(strip, size, d):

    min_val = d
    min_point = []
    for i in range(size):
        j = i + 1
        while j < size and (strip[j][1] -
                            strip[i][1]) < min_val:
            min_val = distance([strip[i], strip[j]])
            min_point = [strip[i],strip[j]]
            j += 1

    return min_val, min_point



def find_closest(points_x, points_y, n):

    if n <= 3:
        return smallest_distance(points_x, n)[0], smallest_distance(points_x, n)[1]

    mid = n // 2
    midPoint = points_x[mid]

    dl = find_closest(points_x[:mid], points_y, mid)[0]
    dr = find_closest(points_x[mid:], points_y, n - mid)[0]
    dl_point = find_closest(points_x[:mid], points_y, mid)[1]
    dr_point = find_closest(points_x[mid:], points_y, n - mid)[1]



    d = min(zip([dl, dr],[dl_point, dr_point]), key = lambda t: t[0])

    strip = []
    for i in range(n):
        if abs(points_y[i][0] - midPoint[0]) < d[0]:
            strip.append(points_y[i])

    return min(zip([d[0], strip_closest(strip, len(strip), d[0])[0]], [d[1],strip_closest(strip, len(strip), d[0])[1]])  )


def closest(points_x, n):
    points_x.sort(key = lambda point: point[0])
    points_y = copy.deepcopy(points_x)
    points_y.sort(key = lambda point: point[1])

    return find_closest(points_x, points_y, n)


# You can change the following points to desired ones and then run the code to see the result
Points = [(2, 3, 4), (7, 2, 5), (4, 6, 1), (9, 7, 10)]
length = len(Points)

print("The closest pair of points are: ", closest(Points, length)[1],
      " and the corresponding distance is: ", closest(Points, length)[0])



```
